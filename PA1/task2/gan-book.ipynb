```python
import cv2
import numpy as np
import torch
import torchvision.transforms.functional as TF # Use torchvision's powerful resize
import torchvision.utils as vutils
import matplotlib.pyplot as plt

# --- Your existing code to generate images ---
checkpoint_path = "checkpoints/last_checkpoint.pth"
checkpoint = torch.load(checkpoint_path, map_location=device)
G.load_state_dict(checkpoint["generator_state_dict"])
G.to(device)
G.eval()
with torch.no_grad():
    noise = torch.randn(20, latent_dim, device=device)
    fake_images = G(noise).cpu()

# Rescale to [0,1] for processing
imgs = (fake_images + 1) / 2
TARGET_SIZE = (56, 56) 
resized_tensors = []

for img_tensor in imgs:
    img_np = (img_tensor.permute(1, 2, 0).numpy() * 255).astype(np.uint8)
    gray = cv2.cvtColor(img_np, cv2.COLOR_RGB2GRAY)
    _, thresh = cv2.threshold(gray, 1, 255, cv2.THRESH_BINARY)
    coords = cv2.findNonZero(thresh)
    if coords is None:
        continue
        
    x, y, w, h = cv2.boundingRect(coords)
    cropped_np = img_np[y:y+h, x:x+w]
    cropped_tensor = torch.from_numpy(cropped_np).permute(2, 0, 1)
    resized_tensor = TF.resize(cropped_tensor, TARGET_SIZE)
    resized_tensors.append(resized_tensor.float() / 255.0)


if resized_tensors:
    # Create the grid from the resized tensors
    grid = vutils.make_grid(
        resized_tensors,
        nrow=4, 
        padding=2  
    )
    grid_height = grid.shape[1]
    grid_width = grid.shape[2]
    
    plt.figure(figsize=(10, 10 * grid_height / grid_width)) 
    plt.imshow(grid.permute(1, 2, 0))
    plt.axis("off")
    plt.subplots_adjust(left=0, right=1, top=1, bottom=0) 
    plt.show()
```

### FID

```python
from torchmetrics.image.fid import FrechetInceptionDistance
import torchvision.transforms.functional as TF

fid = FrechetInceptionDistance(feature=2048).to(device)

for images, _ in val_loader:
    images = TF.resize(images, (299, 299)).to(device)
    images_uint8 = ((images + 1) / 2 * 255).clamp(0, 255).to(torch.uint8)
    fid.update(images_uint8, real=True)

G.eval()
with torch.no_grad():
    total_samples = 10000
    batch_size = 128
    generated = 0
    while generated < total_samples:
        noise = torch.randn(batch_size, latent_dim, device=device)
        fake_images = G(noise)
        fake_images = TF.resize(fake_images, (299, 299))
        fake_uint8 = ((fake_images + 1) / 2 * 255).clamp(0, 255).to(torch.uint8)
        fid.update(fake_uint8.to(device), real=False)
        generated += batch_size

fid_score = fid.compute()
print(f"FID (GAN vs Real): {fid_score.item():.4f}")
```

### Inception Score

```python
from torchvision.models import inception_v3
import torch.nn.functional as F

inception = inception_v3(pretrained=True, transform_input=False).to(device).eval()

def inception_score(samples, splits=10):
    import numpy as np
    N = samples.size(0)
    batch_size = 32
    preds = []
    with torch.no_grad():
        for i in range(0, N, batch_size):
            batch = samples[i:i+batch_size]
            batch = F.interpolate(batch, size=(299, 299), mode='bilinear')
            logits = inception(batch)
            p_yx = F.softmax(logits, dim=1)
            preds.append(p_yx.cpu())
    preds = torch.cat(preds, dim=0).numpy()
    split_scores = []
    for k in range(splits):
        part = preds[k * (N // splits):(k+1)*(N//splits), :]
        py = np.mean(part, axis=0)
        scores = part * (np.log(part) - np.log(py))
        split_scores.append(np.exp(np.mean(np.sum(scores, axis=1))))
    return float(np.mean(split_scores)), float(np.std(split_scores))

# Example
G.eval()
with torch.no_grad():
    z = torch.randn(1000, latent_dim, device=device)
    fake_images = G(z)
is_mean, is_std = inception_score(fake_images)
print("GAN Inception Score:", is_mean, "+/-", is_std)
```

### Interpolation

```python
import cv2
import numpy as np
import torch
import torch.nn.functional as F
import torchvision.transforms.functional as TF
import matplotlib.pyplot as plt

# --- Helper Function (Unchanged, still necessary) ---
def crop_and_resize_tensor(img_tensor, target_size=(48, 48)):
    """Crops black borders and resizes a single image tensor."""
    img_np = (img_tensor.permute(1, 2, 0).numpy() * 255).astype(np.uint8)
    gray = cv2.cvtColor(img_np, cv2.COLOR_RGB2GRAY)
    _, thresh = cv2.threshold(gray, 1, 255, cv2.THRESH_BINARY)
    coords = cv2.findNonZero(thresh)
    if coords is None:
        cropped_tensor = img_tensor.unsqueeze(0)
    else:
        x, y, w, h = cv2.boundingRect(coords)
        cropped_np = img_np[y:y+h, x:x+w]
        cropped_tensor = torch.from_numpy(cropped_np).permute(2, 0, 1).unsqueeze(0)
    resized_tensor = TF.resize(cropped_tensor, target_size).squeeze(0)
    return resized_tensor.float() / 255.0

# --- FINAL VISUALIZATION LOGIC ---
G.eval()
num_steps = 7   # Number of images per interpolation
examples = 3    # Number of rows
TARGET_SIZE = (50, 50)

# 1. Generate all the image data first
all_examples_images = []
for i in range(examples):
    z1 = torch.randn(1, latent_dim, device=device)
    z2 = torch.randn(1, latent_dim, device=device)
    
    interpolated_z = torch.cat([
        z1 * (1 - alpha) + z2 * alpha for alpha in torch.linspace(0, 1, num_steps)
    ]).to(device)
    
    with torch.no_grad():
        interpolated_imgs_raw = G(interpolated_z)
    
    interpolated_imgs_normalized = torch.clamp((interpolated_imgs_raw.cpu() + 1) / 2, 0, 1)
    
    processed_images = [
        crop_and_resize_tensor(img_t, TARGET_SIZE) for img_t in interpolated_imgs_normalized
    ]
    all_examples_images.append(processed_images)

# 2. Create the Matplotlib figure and subplots
# We make the figure size dynamic based on the number of images to get a good aspect ratio
fig, axes = plt.subplots(examples, num_steps, figsize=(num_steps * 1.5, examples * 1.5))

fig.suptitle("GAN Latent Space Interpolations", fontsize=16)

# 3. Populate the subplots with images
for i in range(examples):
    for j in range(num_steps):
        ax = axes[i, j]
        img_tensor = all_examples_images[i][j]
        ax.imshow(img_tensor.permute(1, 2, 0))
        ax.axis('off') # Turn off axis ticks and labels
        
        # Add titles only to the top row
        if i == 0:
            if j == 0:
                ax.set_title("Start")
            elif j == num_steps - 1:
                ax.set_title("End")

# 4. (THE FIX) Manually remove all whitespace
# This forces the subplots to be flush against each other
plt.subplots_adjust(
    left=0.01,    # The left side of the subplots of the figure
    right=0.99,   # The right side of the subplots of the figure
    bottom=0.01,  # The bottom of the subplots of the figure
    top=0.93,     # The top of the subplots of the figure (leaving space for suptitle)
    wspace=0.05,  # The amount of width reserved for space between subplots
    hspace=0.05   # The amount of height reserved for space between subplots
)

plt.show()
```

### Latent Traversal

```python
import torch
import matplotlib.pyplot as plt

def gan_latent_traversal_grid(generator, latent_dim, dim_indices, steps=7, span=3.0, device="cuda"):
    generator.eval()
    results = {}
    base_z = torch.randn(1, latent_dim, device=device)  # one fixed latent vector
    with torch.no_grad():
        values = torch.linspace(-span, span, steps)
        for dim_idx in dim_indices:
            decoded = []
            for v in values:
                z_mod = base_z.clone()
                z_mod[0, dim_idx] += v
                img = generator(z_mod).cpu()
                decoded.append(img[0])  # (C,H,W)
            results[dim_idx] = (values, torch.stack(decoded, dim=0))  # (steps, C,H,W)
    return results

# Example usage
dim_indices = [5, 30, 100]  
decoded_results = gan_latent_traversal_grid(G, latent_dim, dim_indices, steps=5, span=3.0, device=device)

# -------------------------------
# Plot: rows=dim, cols=steps
# -------------------------------
rows = len(dim_indices)
cols = next(iter(decoded_results.values()))[1].size(0)

fig, axes = plt.subplots(rows, cols, figsize=(3*cols, 3*rows))

if rows == 1:
    axes = [axes]

for r, dim_idx in enumerate(dim_indices):
    values, imgs = decoded_results[dim_idx]
    for c in range(cols):
        ax = axes[r][c] if rows > 1 else axes[c]
        ax.imshow(((imgs[c].permute(1, 2, 0) + 1) / 2).clamp(0, 1))
        ax.axis("off")
        if r == 0:  # top row → column labels
            ax.set_title(f"{values[c]:.2f}", fontsize=10)

    # put "Dim <idx>" text at start of the row (left side of first image)
    axes[r][0].text(-imgs.shape[2]//2, imgs.shape[1]//2,
                    f"Dim {dim_idx}", fontsize=14, va="center", ha="right")

plt.suptitle("GAN Latent Traversals (Rows = Dimensions, Cols = Values)", fontsize=16)
plt.tight_layout()
plt.show()
```

### OOD

```python
import cv2
import numpy as np
import torch
import torch.nn.functional as F
import torchvision.transforms.functional as TF
import torchvision.utils as vutils
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

device = device if 'device' in globals() else ("cuda" if torch.cuda.is_available() else "cpu")

# --- Helper Function (Unchanged, still necessary) ---
def crop_and_resize_tensor(img_tensor, target_size=(48, 48)):
    """Crops black borders and resizes a single image tensor."""
    # Ensure tensor is on CPU for numpy conversion
    img_tensor = img_tensor.cpu()
    img_np = (img_tensor.permute(1, 2, 0).numpy() * 255).astype(np.uint8)
    gray = cv2.cvtColor(img_np, cv2.COLOR_RGB2GRAY)
    _, thresh = cv2.threshold(gray, 1, 255, cv2.THRESH_BINARY)
    coords = cv2.findNonZero(thresh)
    if coords is None:
        # If image is all black, return a black tensor of target size
        return torch.zeros(3, target_size[0], target_size[1])
    else:
        x, y, w, h = cv2.boundingRect(coords)
        cropped_np = img_np[y:y+h, x:x+w]
        # Convert back to tensor (C, H, W)
        cropped_tensor = torch.from_numpy(cropped_np).permute(2, 0, 1)
    
    # Resize and return as a float tensor in [0, 1] range
    resized_tensor = TF.resize(cropped_tensor, target_size)
    return resized_tensor.float() / 255.0

# --- Original Generator Functions (Unchanged) ---
def gan_latent_extrapolation_grid(generator, latent_dim, scales=[1.0, 2.0, 4.0, 8.0], n_per_scale=8, device="cuda"):
    generator.eval()
    results = {}
    with torch.no_grad():
        for s in scales:
            zs = torch.randn(n_per_scale, latent_dim, device=device) * s
            imgs = generator(zs).cpu()
            results[s] = imgs
    return results

def gan_pca_traversal_grid(generator, latent_dim, pc_indices=[0, 1], n_samples=5000, steps=7, span=3.0, n_seeds=3, device="cuda"):
    generator.eval()
    Z = torch.randn(n_samples, latent_dim).numpy()
    pca = PCA(n_components=max(pc_indices) + 1)
    pca.fit(Z)
    pcs = pca.components_
    base_z = torch.randn(n_seeds, latent_dim, device=device)
    results = {}
    with torch.no_grad():
        for pc_idx in pc_indices:
            rows = []
            for i in range(n_seeds):
                decoded = []
                z0 = base_z[i:i+1].clone()
                values = torch.linspace(-span, span, steps)
                for v in values:
                    z_mod = z0 + (v * torch.from_numpy(pcs[pc_idx]).float().unsqueeze(0).to(device))
                    img = generator(z_mod).cpu()
                    decoded.append(img)
                decoded = torch.cat(decoded, dim=0)
                rows.append(decoded)
            results[pc_idx] = (rows, base_z.cpu(), pcs[pc_idx])
    return results

# --- UNIFIED SETTINGS ---
TARGET_SIZE = (50, 50)

# ===================================================================
# 1. LATENT EXTRAPOLATION VISUALIZATION (FIXED)
# ===================================================================
print("Generating fixed Latent Extrapolation figures...")
scales = [1.0, 2.0, 4.0, 8.0]
gan_results = gan_latent_extrapolation_grid(G, latent_dim, scales=scales, n_per_scale=8, device=device)

for s, imgs in gan_results.items():
    # Normalize the batch of raw images
    imgs_normalized = ((imgs + 1) / 2).clamp(0, 1)
    
    # Process each image in the batch to crop and resize
    processed_images = [crop_and_resize_tensor(img, TARGET_SIZE) for img in imgs_normalized]
    
    # Create a grid from the clean, uniform-sized images
    grid = vutils.make_grid(processed_images, nrow=8, padding=1)
    
    # Plot the grid tightly
    grid_height = grid.shape[1]
    grid_width = grid.shape[2]
    plt.figure(figsize=(12, 12 * grid_height / grid_width))
    plt.imshow(grid.permute(1, 2, 0))
    plt.axis("off")
    plt.title(f"GAN Samples from N(0, {s:.1f}²)")
    plt.subplots_adjust(left=0, right=1, top=0.95, bottom=0) # Leave a little space for title
    plt.show()

# ===================================================================
# 2. PCA TRAVERSAL VISUALIZATION (FIXED)
# ===================================================================
print("\nGenerating fixed PCA Traversal figures...")
pca_results = gan_pca_traversal_grid(G, latent_dim, pc_indices=[0, 1], n_samples=2000, steps=7, span=2.5, n_seeds=3, device=device)

for pc_idx, (rows, seeds, pc_vec) in pca_results.items():
    # Get dimensions for subplot grid
    num_rows = len(rows)
    num_cols = rows[0].size(0)
    
    # Create the figure
    fig, axes = plt.subplots(num_rows, num_cols, figsize=(num_cols * 1.5, num_rows * 1.5))
    fig.suptitle(f"GAN PCA Traversal — Principal Component {pc_idx}", fontsize=16)

    for r in range(num_rows):
        for c in range(num_cols):
            ax = axes[r, c]
            
            # Get the raw image, normalize it, then crop and resize it
            raw_img = rows[r][c]
            norm_img = ((raw_img + 1) / 2).clamp(0, 1)
            processed_img = crop_and_resize_tensor(norm_img, TARGET_SIZE)
            
            # Display the final, clean image
            ax.imshow(processed_img.permute(1, 2, 0))
            ax.axis("off")

    # Use tight_layout to manage spacing automatically and cleanly
    plt.tight_layout(rect=[0, 0, 1, 0.96]) # rect leaves space for the suptitle
    plt.show()
```

### Model Collapse Check

```python
import torch
import torchvision
import torchvision.transforms as T
import torchvision.models as models
import numpy as np
from sklearn.neighbors import NearestNeighbors
from torch.utils.data import DataLoader
from tqdm import tqdm
import matplotlib.pyplot as plt

# -----------------------------
# Feature extractor (ResNet18)
# -----------------------------
feature_net = models.resnet18(weights=models.ResNet18_Weights.IMAGENET1K_V1)
feature_net.fc = torch.nn.Identity()
feature_net = feature_net.to(device)
feature_net.eval()

# -----------------------------
# Transform for real dataset
# -----------------------------
transform = T.Compose([
    T.Resize(224),
    T.ToTensor(),
    T.Normalize(mean=[0.485, 0.456, 0.406],
                std=[0.229, 0.224, 0.225]),
])

trainset = torchvision.datasets.CIFAR10(root="./data", train=True, download=True, transform=transform)
trainloader = DataLoader(trainset, batch_size=128, shuffle=False, num_workers=4)

# -----------------------------
# Extract real CIFAR-10 features
# -----------------------------
all_feats = []
all_labels = []
with torch.no_grad():
    for imgs, labels in tqdm(trainloader, desc="Extracting CIFAR-10 features"):
        feats = feature_net(imgs.to(device)).cpu().numpy()
        all_feats.append(feats)
        all_labels.append(labels.numpy())
all_feats = np.concatenate(all_feats, axis=0)
all_labels = np.concatenate(all_labels, axis=0)

# -----------------------------
# Fit kNN on real CIFAR-10
# -----------------------------
nn = NearestNeighbors(n_neighbors=1, algorithm="auto").fit(all_feats)

# -----------------------------
# Generate GAN samples
# -----------------------------
with torch.no_grad():
    z = torch.randn(2000, latent_dim, device=device)  # generate 2k samples
    fake_images = G(z).cpu()

# normalize to [0,1] for ResNet input
imgs = (fake_images + 1) / 2

# -----------------------------
# Assign GAN images to nearest CIFAR-10 class
# -----------------------------
gan_loader = DataLoader(imgs, batch_size=64, shuffle=False)

all_assigned = []
with torch.no_grad():
    for batch in tqdm(gan_loader, desc="Classifying GAN samples"):
        imgs_in = T.Resize(224)(batch)
        imgs_in = T.Normalize(mean=[0.485, 0.456, 0.406],
                              std=[0.229, 0.224, 0.225])(imgs_in)
        feats = feature_net(imgs_in.to(device)).cpu().numpy()
        idxs, dists = nn.kneighbors(feats, return_distance=True)
        idxs = idxs.astype(int)  # FIXED: cast to int
        assigned = all_labels[idxs[:, 0]]
        all_assigned.append(assigned)
all_assigned = np.concatenate(all_assigned, axis=0)

# -----------------------------
# Count frequency of assigned classes
# -----------------------------
unique, counts = np.unique(all_assigned, return_counts=True)
freq = dict(zip(unique, counts))
print("Class frequencies among GAN samples:", freq)

# -----------------------------
# Plot histogram
# -----------------------------
plt.figure(figsize=(8, 4))
plt.bar([trainset.classes[i] for i in unique], counts)
plt.xticks(rotation=45)
plt.title("GAN generated samples - nearest CIFAR-10 class distribution")
plt.show()
```
